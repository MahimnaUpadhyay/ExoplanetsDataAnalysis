```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import seaborn as sns
sns.set()
```

```python
dataset = pd.read_csv('D:/CODING/Data Analysts/Data_Sets/Exoplanents/ImprovedData.csv')
```

```python
dataset
```

```python
neptune_like = dataset[dataset['planet_type'] == 'Neptune-like']
gas_giants = dataset[dataset['planet_type'] == 'Gas Giant']
super_earth = dataset[dataset['planet_type'] == 'Super Earth']
terrestrial = dataset[dataset['planet_type'] == 'Terrestrial']
```

```python
plt.figure(figsize = (20,10))
plt.title('Exoplanets Near Us')

plt.scatter(neptune_like['distance'], neptune_like['planet_type'], label='Neptune-like', color='blue')
plt.scatter(gas_giants['distance'], gas_giants['planet_type'], label='Gas Giants', color='red')
plt.scatter(super_earth['distance'], super_earth['planet_type'], label='Super Earth', color='yellow')
plt.scatter(terrestrial['distance'], terrestrial['planet_type'], label='Terrestrial', color='green')

plt.xlabel('Distance (in light years)')
plt.ylabel('Planet Type')
plt.legend()

plt.show()
```

```python
# Percentage of Planet Types
```

```python
planet_type_counts = dataset['planet_type'].value_counts()
```

```python
plt.figure(figsize=(10, 8))
plt.title('Planet Type Ratio')

plt.pie(planet_type_counts, 
        labels=planet_type_counts.index, 
        autopct='%1.1f%%',
       textprops={
           'size':'xx-large',
           'color' : 'w'
       })

plt.legend(loc='lower left')
plt.show()
```

```python
# Bar Chart showing Detection Method for Terrestrial Planets
```

```python
plt.figure(figsize=(20,6))
plt.title('Detection Method used for Terrestrial Planets', fontsize='20', fontweight='bold')

plt.bar(x=dataset['detection_method'], 
        height=dataset['planet_type']=='Terrestrial', 
        width=0.25)

plt.xticks(rotation=-90)
plt.xlabel('Detection Methods', fontsize='15')
plt.ylabel('Terrestrial Planets', fontsize='15')

plt.show()
```
